### [실습] 다양한 방법으로 배열 생성


NumPy를 사용하여 다음과 같은 배열을 생성해 보세요.

1. 크기가 4x3인 모든 값이 0인 배열
2. 크기가 3x3인 모든 값이 1인 배열
3. 0에서 20까지 5씩 증가하는 값으로 구성된 배열
4. 0과 1 사이의 4개의 등간격 값을 가진 배열
5. *크기가 2x2인 정규 분포로 구성된 배열*

```python
import numpy as np

# 1. 크기가 4x3인 모든 값이 0인 배열
zeros_array = np.zeros((4,3), dtype=int)
print("zeros_array: \n", zeros_array)

# 2. 크기가 3x3인 모든 값이 1인 배열
ones_array = np.ones((3,3), dtype=int)
print("ones_array: \n", ones_array)

# 3. 0에서 20까지 5씩 증가하는 값으로 구성된 배열
array = np.arange(0, 20, 5)
print("array: \n", array)

# 4. 0과 1 사이의 4개의 동간격 값을 가진 배열
split_array = np.linspace(0, 1, 4)
print("split_array: \n", split_array)

# 5. 크기가 2x2인 정규 분포로 구성된 배열
reg_array = np.random.normal(10, 1, (2,2))
print("reg_array: \n", reg_array)
```
