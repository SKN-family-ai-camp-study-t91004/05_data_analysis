다음은 5개 지점에서 판매된 과일에 대한 월별 판매량 데이터입니다. 이 데이터는 각 지점의 판매량을 월별로 나타내며, 각 지점에서 판매된 사과, 바나나, 오렌지의 판매량이 기록되어 있습니다. 이 데이터를 사용하여 여러 가지 통계적 분석과 배열 조작을 수행하세요.

1. 각 지점에서 6개월 동안 판매된 **사과**, **바나나**, **오렌지**의 월별 평균 판매량을 계산하세요.
2. 각 월마다 전체 지점에서 가장 많이 판매된 과일을 찾아 판매량과 과일명을 출력하세요.
3. 6개월 동안 전체 지점에서 가장 많이 판매된 과일의 총량을 과일별로 구하세요. 어떤 과일이 가장 많이 판매되었는지 확인하세요.
4. 각 지점에서 판매량이 150개 이상인 데이터를 추출하고 그에 해당하는 지점과 과일을 출력하세요.

- **데이터 (**각 지점의 월별 과일 판매량(단위: 개))

    ```python
    import numpy as np

    # 5개 지점의 월별 사과, 바나나, 오렌지 판매량 (5x3x6 배열)
    # 5개의 지점, 3개의 과일(사과, 바나나, 오렌지), 6개월치 데이터
    sales_data = np.array([
        # 지점 1
        [[120, 150, 200, 230, 180, 210],  # 사과 판매량
         [100, 140, 170, 190, 160, 180],  # 바나나 판매량
    ```

````python
import numpy as np

# 5개 지점의 월별 사과, 바나나, 오렌지 판매량 (5x3x6 배열)
# 5개의 지점, 3개의 과일(사과, 바나나, 오렌지), 6개월치 데이터
sales_data = np.array([
    # 지점 1
    [[120, 150, 200, 230, 180, 210],  # 사과 판매량
     [100, 140, 170, 190, 160, 180],  # 바나나 판매량
     [90, 120, 160, 140, 130, 150]],  # 오렌지 판매량

    # 지점 2
    [[80, 90, 110, 140, 130, 120],
     [60, 70, 90, 100, 110, 95],
     [50, 80, 100, 110, 90, 85]],

    # 지점 3
    [[130, 140, 160, 180, 170, 190],
     [110, 120, 130, 150, 140, 160],
     [100, 110, 120, 130, 120, 140]],

    # 지점 4
    [[100, 130, 150, 170, 160, 180],
     [90, 100, 110, 120, 130, 125],
     [80, 110, 140, 130, 120, 110]],

    # 지점 5
    [[90, 110, 130, 140, 120, 130],
     [70, 85, 100, 110, 105, 95],
     [60, 90, 110, 100, 90, 95]]
])

# 1. 각 지점에서 6개월 동안 판매된 사과, 바나나, 오렌지의 월별 평균 판매량을 계산하세요.
result = np.mean(sales_data, axis=2)
print("지점1 과일별 월별 평균 판매량: ", result[0])
print("지점2 과일별 월별 평균 판매량: ", result[1])
print("지점3 과일별 월별 평균 판매량: ", result[2])
print("지점4 과일별 월별 평균 판매량: ", result[3])
print("지점5 과일별 월별 평균 판매량: ", result[4])

print()
# 2. 각 월마다 전체 지점에서 가장 많이 판매된 과일을 찾아 판매량과 과일명을 출력하세요.
fruits = ['사과', '바나나', '오렌지']
months = ['1월', '2월', '3월', '4월', '5월', '6월']

total = np.sum(sales_data, axis=0)
best_fruit_idx = np.argmax(total, axis=0) # 각 지점 월마다 과일별 합계 중 최대값 인덱스 반환(argmax)
best_sales = np.max(total, axis=0) # 각 지점 월마다 제일 많이 팔린 판매량 반환

for i in range(6):
    print(f"{months[i]} : {fruits[best_fruit_idx[i]]}, {best_sales[i]}개")

print()
# 3. 6개월 동안 전체 지점에서 가장 많이 판매된 과일의 총량을 구하세요. 어떤 과일이 가장 많이 판매되었는지 확인하세요.
total = np.sum(sales_data, axis=(0,2))
best_fruit_idx = np.argmax(total)

print(f"가장 많이 판매된 과일 : {fruits[best_fruit_idx]}, 그 판매량 : {total[best_fruit_idx]}개")

print()
# 4. 각 지점에서 판매량이 150개 이상인 데이터를 추출하고 그에 해당하는 지점과 과일을 출력하세요.
result = set()

for g in range(5):
    for f in range(3):
        for m in range(6):
            if sales_data[g, f, m] >= 150:
                result.add((g+1, fruits[f]))

for g, f in sorted(result):
    print(f"지점{g} / {f}")
````
